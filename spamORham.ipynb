```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('email_classification.csv')
dataset.head()
```

```python
dataset.value_counts()
```

```python
dataset.label.value_counts()
```

```python
dataset['spam'] = dataset['label'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset.email, dataset.spam, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train_cv = cv.fit_transform(X_train.values)
X_test_cv = cv.transform(X_test.values)
```

```python
X_train_cv = X_train_cv
```

```python
X_test_np = X_test_cv
```

```python
print(X_train_cv)
```

```python
from sklearn.naive_bayes import MultinomialNB
cl = MultinomialNB()
cl.fit(X_train_cv, y_train)
```

```python
X_test_cv = cv.fit_transform(X_test)

y_pred = cl.predict(X_test_np)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))

accuracy_score(y_pred,y_test)
```
